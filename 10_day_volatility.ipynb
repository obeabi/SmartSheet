```python
import yfinance as yf
import pandas as pd
import pandas as pd
print("libraries installed!")
```

```python
#List of ETfs
data = pd.read_csv('sector_etfs.csv')
# List of ETFs
etfs = data['Ticker'].tolist()
etfs[0:4]
```

```python
 #Function to calculate 10-day volatility
def calculate_10_day_volatility(ticker):
    # Fetch historical data for the last month
    data = yf.download(ticker, period="1mo", interval="1d")

    # Ensure we have at least 10 rows
    if data.shape[0] < 10:
        raise ValueError(f"Not enough data to calculate 10-day volatility for {ticker}")

    # Calculate daily returns
    data['Daily Return'] = data['Close'].pct_change()

    # Calculate 10-day rolling volatility
    data['10-Day Volatility'] = data['Daily Return'].rolling(window=10).std()

    # Get the most recent 10-day volatility
    latest_volatility = data['10-Day Volatility'].iloc[-1]

    return latest_volatility
```

```python
# Calculate and store 10-day volatility for each ETF
volatility_data = {}
for etf in etfs:
    try:
        volatility = calculate_10_day_volatility(etf)
        volatility_data[etf] = volatility
    except Exception as e:
        volatility_data[etf] = f"Error: {e}"

# Convert to DataFrame for better visualization
volatility_df = pd.DataFrame(list(volatility_data.items()), columns=['ETF', '10-Day Volatility'])

# Display the results
volatility_df.to_csv('sector_10day_volatitly_data.csv', index=False)

volatility_df.head()
```
